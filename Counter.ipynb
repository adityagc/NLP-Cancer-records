```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
```

```python
train = pd.read_csv("train_cleaned_text.csv", encoding = "ISO-8859-1")
```

```python
train.head()
```

```python
concat = train['TEXT_PATH_CLINICAL_HISTORY'].astype(str) + " " + train['TEXT_PATH_COMMENTS'].astype(str) + " " + train['TEXT_PATH_FORMAL_DX'].astype(str)+ " " + train['TEXT_PATH_FULL_TEXT'].astype(str)+ " " + train['TEXT_PATH_GROSS_PATHOLOGY'].astype(str)+ " " + train['TEXT_PATH_MICROSCOPIC_DESC'].astype(str)+ " " + train['TEXT_PATH_NATURE_OF_SPECIMENS'].astype(str)+ " " + train['TEXT_PATH_STAGING_PARAMS'].astype(str)+ " " + train['TEXT_PATH_SUPP_REPORTS_ADDENDA'].astype(str)
new = pd.DataFrame()
new['SITE'] = train['SITE']
new['PATIENT_DISPLAY_ID']= train['PATIENT_DISPLAY_ID']
new['Text'] = concat
```

```python
new.Text[18]
```

```python
new['Text'] = new.Text.str.lower()
```

```python
text = new.Text.str.split(' ')
```

```python
colon_count = []
breast_count = []
lung_count = []
prostate_count = []
other_count= []
for line in text:
    colcount = 0
    brcount = 0
    lucount = 0
    prcount = 0
    for word in line:
        if (word == 'colon'):
            colcount = colcount + 1
        if (word == 'breast'):
            brcount = brcount + 1
        if (word == 'lung'):
            lucount = lucount + 1
        if (word == 'prostate'):
            prcount = prcount + 1
    colon_count.append(colcount)
    breast_count.append(brcount)
    lung_count.append(lucount)
    prostate_count.append(prcount)
```

```python
new['Other_count'] = 0
new['Colon_count'] = colon_count
new['lung_count'] = lung_count
new['breast_count'] = breast_count
new['prostate_count'] = prostate_count
```

```python
new.head()
```

```python
nums = pd.DataFrame()
nums['Other_count'] = new['Other_count']
nums['Colon_count'] = colon_count
nums['lung_count'] = lung_count
nums['breast_count'] = breast_count
nums['prostate_count'] = prostate_count
```

```python
nums
```

```python
arr = nums.values
argmaxvals  = np.argmax(arr, axis=1)
dec_list =['OTHER','COLON','LUNG','BREAST','PROSTATE']
dec_series = []
for i in range(np.shape(argmaxvals)[0]):
    dec_series.append(dec_list[argmaxvals[i]])
```

```python
new['Prediction'] = dec_series
```
